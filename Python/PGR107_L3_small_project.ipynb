```
# P3.2.1 Two variables a and b, implement a function called compareTwoVariables(a,b) using if-elif-else to determine if a is less or equal or larger than b.
# P3.2.2 Assign a=10, b=17, call the function compareTwoVariables() to get correct output information
# P3.2.3 assign random values to variables x , y and z. Find the largest number among the three variables by calling the function compareTwoVariables().

#imports the random method for x, y and z
import random

#Sets a and b
a = 10
b = 17

#Generates a random number between 1 and 666
x = random.randint(1, 666)
y = random.randint(1, 666)
z = random.randint(1, 666)

#Compare two arguments
def compareTwoVariables(a, b):
    #Checks if a is bigger than b
    if a>b:
        check=a
        return check
    #Checks if b is bigger than a
    elif b>a:
        check=b
        return check
    #Checks if a and b are equal
    elif a==b:
        check=a
        return check
    #Prints if none of the checks above are valid
    else:
        print("Something went wrong!")


#Calls compareTwoVariables with two arguments, a and b
print("The first numbers to be compared are", a, "and", b)
print("The greatest number is", compareTwoVariables(a, b))

#Calls compareTwoVariables with three arguments, x, y and z
#First, compareTwoVariables is called with x and y
compareOne=compareTwoVariables(x, y)
print("\nThe randomly generated numbers to be compared are", x, y, z)
#Secondly, compareTwoVariables is called with the biggest number from compareOne above and z
compareTwo=compareTwoVariables(compareOne, z)
#The biggest number from x y z is printed below
print("The biggest number is", compareTwo)
```
